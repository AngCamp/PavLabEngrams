```r
# exonic read (spliced) expression matrix
emat <- jeager_dat$spliced;
# intronic read (unspliced) expression matrix
nmat <- jeager_dat$unspliced
# spanning read (intron+exon) expression matrix
smat <- jeager_dat$spanning;
```
